```python
###want to create 13 lists
# [[],[],[],[]....]
allElements=[]
for i in range(13):
    allElements.append([])
    
```

```python
data['id;age;gender;height;weight;ap_hi;ap_lo;cholesterol;gluc;smoke;alco;active;cardio'].head(2)
    
```

```python
### add this data to all_element and next we will make it ready
for i in data['id;age;gender;height;weight;ap_hi;ap_lo;cholesterol;gluc;smoke;alco;active;cardio']:
    splitedEle=i.split(';')
    for ele in range(len(splitedEle)):
        allElements[ele].append(splitedEle[ele])
```

```python
df=pd.DataFrame()
for i in range(13):
    df[listOfColumns[i]]=allElements[i]
    
```

```python
for i in ["id", "age", "gender", "height", "weight", "ap_hi", "ap_lo", "cholesterol", "gluc", "smoke", "alco", "active", "cardio"]:
    if i =="weight":
        df[i]=[int(float(x)) for x in df[i]]
        break
    df[i]=[int(x) for x in df[i]]
# df['age']=[int(x) for x in df['age']]
```

```python
### Cleand Data
```

```python
df=pd.read_csv("../data/clean_data.csv")
```

```python
df.head()
```

```python
df_copy=df
```

```python
df_copy.drop(columns=["Unnamed: 0","id"],inplace=True)
```

```python
df.to_csv("data/clean_data.csv",index=False)
```

```python
df_copy.head(2)
```

```python
X=df_copy.drop("cardio",axis=1)
y=df_copy["cardio"]
```

```python
### spliting the Data 
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report

print("accuracy_score ",accuracy_score(y_test,y_pred))
print("classification_report is ",classification_report(y_test,y_pred))
```

### we got 62% accuracy with DecisionTreeClassifier

### lets do HyperParameter for this

```python
hyperPara={
    "criterion":["gini", "entropy", "log_loss"],
    "splitter":["best", "random"],
    "max_depth":[None,2,4,6,8,10],
    
}
```

```python
from sklearn.model_selection import GridSearchCV

gridSearch=GridSearchCV(DecisionTreeClassifier(),hyperPara,cv=5)

gridSearch.fit(X_train,y_train)
```

```python
bestParams = gridSearch.best_params_
bestScore = gridSearch.best_score_
print("Best Parameters ", bestParams)
print("Best Cross-Validation Score ", bestScore)
```

### Lets now use best parameter

```python
clfPara=DecisionTreeClassifier(**bestParams)
clfPara.fit(X_train,y_train)
```

```python
y_para_pred=clfPara.predict(X_test)
```

```python
print("Accuracy Score is ",accuracy_score(y_test,y_para_pred))
```

```python
print(classification_report(y_test, y_para_pred))
```

### Improved the accuracy by 10%
